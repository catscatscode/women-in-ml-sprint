**B. Changing the number of features - dropping some columns**

Let's drop some features and see how it affects our regression model's fit.

`Age` and `years of experience` are likely to be two features that would be highly correlated with each other. Keeping them both is probably a bit redundant. Let's try dropping `age` and re-running the model. The r-squared is not expected to change much (from 0.43754799942321354).

*Again, here is a reminder of the columns in the table:* `['Gender', 'Age', 'Department_code', 'Years_exp', 'Tenure (months)']`

```python
features = salaries.drop(['Gross', 'Department', 'Age'], axis ='columns')
target = salaries['Gross']
model = LinearRegression()
model.fit(features, target)
model.score(features, target)
```

That r-squared is indeed nearly identical! 
It means that the new model has one less feature, but is still able to explain just as much variation in the salary.

Now let's try to remove another column, this time something that is expected to affect the model score a bit more.

For example, we can drop *both* `age` and `years of experience`. That's taking away a lot of information from our salary prediction efforts, and is more likely to hurt the model's score.

```python
features = salaries.drop(['Gross', 'Department', 'Age', 'Years_exp'], axis ='columns')
target = salaries['Gross']
model = LinearRegression()
model.fit(features, target)
model.score(features, target)
```

As expected, the model's score dropped by more than half!

And just one more experiment. Since years of experience looks to be an important feature, this time let's drop *all* features *except* `years of experience` and see what happens to the model score.

```python
features = salaries.drop(['Gross', 'Department', 'Gender', 'Age', 'Department_code', 'Tenure (months)'], axis ='columns')
target = salaries['Gross']
model = LinearRegression()
model.fit(features, target)
model.score(features, target)
```

This model has just one single feature, and it explains nearly as much variation in salaries (42%) as the first model could (43.8%) with 5 features! Looks like experience is a very important factor in this company's compensation practices.
